````
# 🤖 AI Collaboration: Data Loading and Cleaning Function
# Ask AI to help you create a robust data loading function

def load_and_clean_transaction_data(file_path):
    """
    Load and clean transaction data for the Smart Finance Assistant

    🤖 AI Collaboration Prompt:
    "Create a function to load CSV transaction data with Date, Amount, Category,
    Description columns. Handle dollar signs in Amount, missing values, and
    data validation. Include clear business-focused error messages."

    Args:
        file_path: Path to CSV file, file object, or pandas DataFrame
    Returns:
        pandas.DataFrame: Cleaned transaction data
    """
    import os
    import pandas as pd
    import numpy as np
    from pandas.errors import EmptyDataError

    if file_path is None:
        raise ValueError("Please provide a transaction CSV file or a pandas DataFrame.")

    try:
        if isinstance(file_path, pd.DataFrame):
            df = file_path.copy()
        else:
            csv_source = (
                file_path.name
                if hasattr(file_path, "name") and not isinstance(file_path, (str, bytes, os.PathLike))
                else file_path
            )
            df = pd.read_csv(csv_source)
    except EmptyDataError:
        raise ValueError(
            "The transaction file is empty. Please upload a CSV with Date, Amount, Category and Description columns."
        )
    except FileNotFoundError:
        raise ValueError(
            "The transaction file could not be found. Please check the file path or upload the CSV again."
        )
    except Exception as e:
        raise ValueError(
            f"Unable to load the transaction CSV. Please check the file format. Details: {e}"
        )

    if df.empty:
        raise ValueError("The transaction file has no rows to analyse.")

    df.columns = [str(col).strip() for col in df.columns]

    required_columns = ["Date", "Amount", "Category", "Description"]
    column_lookup = {str(col).strip().lower(): col for col in df.columns}
    missing_columns = [col for col in required_columns if col.lower() not in column_lookup]

    if missing_columns:
        raise ValueError(
            "Missing required transaction column(s): "
            + ", ".join(missing_columns)
            + ". The CSV must include Date, Amount, Category and Description."
        )

    df = df[[column_lookup[col.lower()] for col in required_columns]].copy()
    df.columns = required_columns
    df = df.dropna(how="all")

    if df.empty:
        raise ValueError("The transaction file only contains blank rows.")

    df["Category"] = df["Category"].fillna("Uncategorised").astype(str).str.strip()
    df.loc[df["Category"].isin(["", "nan", "None", "NaN"]), "Category"] = "Uncategorised"

    df["Description"] = df["Description"].fillna("No description provided").astype(str).str.strip()
    df.loc[df["Description"].isin(["", "nan", "None", "NaN"]), "Description"] = "No description provided"

    cleaned_amount = (
        df["Amount"]
        .astype(str)
        .str.strip()
        .str.replace("−", "-", regex=False)
        .str.replace(r"[\$,]", "", regex=True)
        .str.replace(r"^\((.*)\)$", r"-\1", regex=True)
    )
    cleaned_amount = cleaned_amount.replace(["", "nan", "NaN", "None", "none"], pd.NA)

    df["Amount"] = pd.to_numeric(cleaned_amount, errors="coerce")

    invalid_amounts = df["Amount"].isna()
    if invalid_amounts.any():
        bad_rows = (df.index[invalid_amounts] + 2).astype(str).tolist()
        raise ValueError(
            "Some transaction amounts could not be read as numbers. "
            f"Please check CSV row(s): {', '.join(bad_rows[:5])}."
        )

    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

    invalid_dates = df["Date"].isna()
    if invalid_dates.any():
        bad_rows = (df.index[invalid_dates] + 2).astype(str).tolist()
        raise ValueError(
            "Some transaction dates could not be read. "
            f"Please check CSV row(s): {', '.join(bad_rows[:5])} and use a format like YYYY-MM-DD."
        )

    df = df.sort_values("Date").reset_index(drop=True)

    df["Transaction_Type"] = np.select(
        [df["Amount"] > 0, df["Amount"] < 0],
        ["Expense", "Refund"],
        default="Zero"
    )

    df["Month"] = df["Date"].dt.to_period("M").astype(str)

    return df

# Test your function
clean_data = load_and_clean_transaction_data(df_sample)
print(clean_data)
````

````
# 🤖 AI Collaboration: Spending Analysis Function
# Ask AI to help you create comprehensive spending analysis

def analyze_spending_patterns(df):
    """
    Analyze spending patterns and generate business insights

    🤖 AI Collaboration Prompt:
    "Create a function that analyzes spending by category, calculates percentages,
    identifies top spending areas, and generates actionable financial insights
    formatted for business presentation."

    Args:
        df: Cleaned transaction DataFrame
    Returns:
        dict: Analysis results and insights
    """
    import pandas as pd

    if not isinstance(df, pd.DataFrame):
        raise ValueError(
            "Spending analysis needs a pandas DataFrame created by load_and_clean_transaction_data()."
        )

    required_columns = {"Date", "Amount", "Category", "Description"}
    missing_columns = required_columns - set(df.columns)

    if missing_columns:
        raise ValueError(
            f"Cannot analyse spending because these columns are missing: {', '.join(sorted(missing_columns))}."
        )

    if df.empty:
        raise ValueError("There are no transactions to analyse.")

    working_df = df.copy()
    working_df["Amount"] = pd.to_numeric(working_df["Amount"], errors="coerce")

    if working_df["Amount"].isna().any():
        raise ValueError("The Amount column contains invalid values. Please clean the data before analysis.")

    working_df["Date"] = pd.to_datetime(working_df["Date"], errors="coerce")

    if working_df["Date"].isna().any():
        raise ValueError("The Date column contains invalid values. Please clean the data before analysis.")

    expenses = working_df[working_df["Amount"] > 0].copy()
    refunds = working_df[working_df["Amount"] < 0].copy()

    total_spending = float(expenses["Amount"].sum())
    total_refunds = float(abs(refunds["Amount"].sum()))
    net_spending = float(working_df["Amount"].sum())

    if total_spending > 0:
        category_summary = (
            expenses.groupby("Category", as_index=False)
            .agg(
                TotalSpent=("Amount", "sum"),
                TransactionCount=("Amount", "count"),
                AverageTransaction=("Amount", "mean")
            )
            .sort_values("TotalSpent", ascending=False)
            .reset_index(drop=True)
        )

        category_summary["Percentage"] = (category_summary["TotalSpent"] / total_spending * 100).round(2)
        category_summary["TotalSpent"] = category_summary["TotalSpent"].round(2)
        category_summary["AverageTransaction"] = category_summary["AverageTransaction"].round(2)
        category_summary["FormattedTotal"] = category_summary["TotalSpent"].map(lambda value: f"${value:,.2f}")

        top_row = category_summary.iloc[0]
        top_category = str(top_row["Category"])
        top_category_amount = float(top_row["TotalSpent"])
        top_category_percentage = float(top_row["Percentage"])

        largest_row = expenses.sort_values("Amount", ascending=False).iloc[0]
        largest_transaction = {
            "Date": largest_row["Date"].strftime("%Y-%m-%d"),
            "Amount": float(largest_row["Amount"]),
            "Category": str(largest_row["Category"]),
            "Description": str(largest_row["Description"])
        }
    else:
        category_summary = pd.DataFrame(
            columns=[
                "Category",
                "TotalSpent",
                "TransactionCount",
                "AverageTransaction",
                "Percentage",
                "FormattedTotal"
            ]
        )
        top_category = None
        top_category_amount = 0.0
        top_category_percentage = 0.0
        largest_transaction = None

    discretionary_categories = [
        "Coffee",
        "Dining",
        "Entertainment",
        "Shopping",
        "Takeaway",
        "Restaurants",
        "Subscriptions"
    ]

    discretionary_spending = float(
        expenses[expenses["Category"].isin(discretionary_categories)]["Amount"].sum()
    )

    discretionary_percentage = (
        discretionary_spending / total_spending * 100
        if total_spending > 0
        else 0.0
    )

    refund_summary = (
        refunds.assign(RefundAmount=refunds["Amount"].abs())
        .groupby("Category", as_index=False)
        .agg(
            TotalRefunds=("RefundAmount", "sum"),
            RefundCount=("RefundAmount", "count")
        )
        .sort_values("TotalRefunds", ascending=False)
        .reset_index(drop=True)
        if not refunds.empty
        else pd.DataFrame(columns=["Category", "TotalRefunds", "RefundCount"])
    )

    date_start = working_df["Date"].min()
    date_end = working_df["Date"].max()
    days_covered = max((date_end - date_start).days + 1, 1)

    business_insights = []

    if total_spending == 0:
        business_insights.append(
            "No positive spending transactions were found; the data appears to contain only refunds or zero-value transactions."
        )
    else:
        business_insights.append(
            f"Your largest spending category is {top_category}, accounting for "
            f"${top_category_amount:,.2f} ({top_category_percentage:.1f}% of spending)."
        )

        if top_category_percentage >= 35:
            business_insights.append(
                f"{top_category} is a major budget driver. A 10% reduction in this category "
                f"could save about ${top_category_amount * 0.10:,.2f} over this period."
            )

        if discretionary_percentage >= 40:
            business_insights.append(
                f"Flexible lifestyle categories such as coffee, dining and entertainment make up "
                f"{discretionary_percentage:.1f}% of spending, so small habit changes could have a noticeable impact."
            )

        coffee_total = (
            float(category_summary.loc[category_summary["Category"].eq("Coffee"), "TotalSpent"].sum())
            if not category_summary.empty
            else 0.0
        )

        if coffee_total > 0:
            business_insights.append(
                f"Coffee spending totals ${coffee_total:,.2f}. "
                "Setting a weekly coffee cap is a simple student-friendly savings action."
            )

        daily_average = total_spending / days_covered
        business_insights.append(
            f"Average daily spending across the period is ${daily_average:,.2f}, "
            f"based on {days_covered} day(s) of transactions."
        )

    if total_refunds > 0:
        business_insights.append(
            f"Refunds total ${total_refunds:,.2f}. "
            "They are tracked separately so they do not hide your real spending habits."
        )

    return {
        "total_spending": round(total_spending, 2),
        "total_refunds": round(total_refunds, 2),
        "net_spending": round(net_spending, 2),
        "transaction_count": int(len(working_df)),
        "expense_transaction_count": int(len(expenses)),
        "refund_transaction_count": int(len(refunds)),
        "average_expense": round(float(expenses["Amount"].mean()), 2) if not expenses.empty else 0.0,
        "category_count": int(expenses["Category"].nunique()) if not expenses.empty else 0,
        "category_summary": category_summary,
        "category_totals": {
            row["Category"]: float(row["TotalSpent"])
            for _, row in category_summary.iterrows()
        },
        "refund_summary": refund_summary,
        "top_category": top_category,
        "top_category_amount": round(top_category_amount, 2),
        "top_category_percentage": round(top_category_percentage, 2),
        "largest_transaction": largest_transaction,
        "discretionary_spending": round(discretionary_spending, 2),
        "discretionary_percentage": round(discretionary_percentage, 2),
        "date_range": {
            "start": date_start.strftime("%Y-%m-%d"),
            "end": date_end.strftime("%Y-%m-%d"),
            "days": int(days_covered)
        },
        "business_insights": business_insights
    }

# Test your function
# Test your function
analysis = analyze_spending_patterns(clean_data)

print("SPENDING ANALYSIS SUMMARY")
print("-" * 50)
print(f"Total Spending: ${analysis['total_spending']:.2f}")
print(f"Total Refunds: ${analysis['total_refunds']:.2f}")
print(f"Net Spending: ${analysis['net_spending']:.2f}")
print(f"Total Transactions: {analysis['transaction_count']}")
print(f"Expense Transactions: {analysis['expense_transaction_count']}")
print(f"Refund Transactions: {analysis['refund_transaction_count']}")
print(f"Average Expense: ${analysis['average_expense']:.2f}")

print("\nCATEGORY BREAKDOWN")
print("-" * 50)
display(analysis["category_summary"])

print("\nREFUND SUMMARY")
print("-" * 50)

if not analysis["refund_summary"].empty:
    display(analysis["refund_summary"])
else:
    print("No refunds found.")

print("\nTOP SPENDING CATEGORY")
print("-" * 50)
print(f"Category: {analysis['top_category']}")
print(f"Amount: ${analysis['top_category_amount']:.2f}")
print(f"Percentage: {analysis['top_category_percentage']:.2f}%")

print("\nLARGEST TRANSACTION")
print("-" * 50)

largest = analysis["largest_transaction"]

if largest is not None:
    print(f"Date: {largest['Date']}")
    print(f"Description: {largest['Description']}")
    print(f"Category: {largest['Category']}")
    print(f"Amount: ${largest['Amount']:.2f}")
else:
    print("No expense transactions found.")

print("\nBUSINESS INSIGHTS")
print("-" * 50)

for insight in analysis["business_insights"]:
    print(f"- {insight}")
````

````
# 🤖 AI Collaboration: Business Insights Generator
# Ask AI to help create professional financial recommendations

def generate_financial_recommendations(analysis_data):
    """
    Generate actionable financial recommendations based on spending analysis

    🤖 AI Collaboration Prompt:
    "Based on spending analysis data, create professional financial
    recommendations. Include specific savings opportunities, spending
    pattern observations, and actionable advice formatted for a
    personal finance app user."

    Args:
        analysis_data: Dictionary with spending analysis results
    Returns:
        str: Formatted recommendations report
    """
    import pandas as pd

    if not isinstance(analysis_data, dict):
        raise ValueError("Recommendations need the dictionary returned by analyze_spending_patterns().")

    total_spending = float(analysis_data.get("total_spending", 0) or 0)
    total_refunds = float(analysis_data.get("total_refunds", 0) or 0)
    net_spending = float(analysis_data.get("net_spending", 0) or 0)

    top_category = analysis_data.get("top_category")
    top_amount = float(analysis_data.get("top_category_amount", 0) or 0)
    top_percentage = float(analysis_data.get("top_category_percentage", 0) or 0)

    date_range = analysis_data.get("date_range", {})
    insights = analysis_data.get("business_insights", [])

    category_summary = analysis_data.get("category_summary", pd.DataFrame())

    if isinstance(category_summary, list):
        category_summary = pd.DataFrame(category_summary)
    elif isinstance(category_summary, dict):
        category_summary = pd.DataFrame(category_summary)
    elif category_summary is None:
        category_summary = pd.DataFrame()

    lines = []
    lines.append("## 💡 Budget Buddy Financial Recommendations")
    lines.append("")
    lines.append(f"**Period analysed:** {date_range.get('start', 'N/A')} to {date_range.get('end', 'N/A')}")
    lines.append(f"**Total spending:** ${total_spending:,.2f}")
    lines.append(f"**Refunds tracked separately:** ${total_refunds:,.2f}")
    lines.append(f"**Net cash impact:** ${net_spending:,.2f}")
    lines.append("")

    if total_spending <= 0:
        lines.append("### Summary")
        lines.append(
            "No positive spending transactions were found, so there is not enough spending data "
            "to make category-based recommendations."
        )
        lines.append("Add a normal month of transaction data to receive more useful budgeting guidance.")
        return "\n".join(lines)

    lines.append("### Key observations")

    if top_category:
        lines.append(
            f"- Your biggest spending area is **{top_category}** at **${top_amount:,.2f}**, "
            f"which is **{top_percentage:.1f}%** of all spending."
        )

    for insight in insights[:4]:
        lines.append(f"- {insight}")

    lines.append("")
    lines.append("### Action plan")

    if not category_summary.empty and {"Category", "TotalSpent", "Percentage"}.issubset(category_summary.columns):
        top_categories = category_summary.head(3)

        for _, row in top_categories.iterrows():
            category = str(row["Category"])
            amount = float(row["TotalSpent"])
            percentage = float(row["Percentage"])
            potential_saving = amount * 0.10

            if category.lower() == "coffee":
                action = (
                    "try a weekly coffee limit, bring coffee from home on two days, "
                    "or use campus loyalty deals"
                )
            elif category.lower() in ["dining", "restaurants", "takeaway"]:
                action = "plan two low-cost meals before buying takeaway and set a weekly eating-out cap"
            elif category.lower() == "entertainment":
                action = (
                    "choose one paid activity per week and mix in free events, "
                    "student discounts, or streaming nights"
                )
            elif category.lower() == "groceries":
                action = "shop with a list, compare unit prices, and plan meals around specials before buying"
            elif category.lower() == "transport":
                action = "compare public transport, walking, car-pooling, and ride-share costs before each trip"
            elif category.lower() == "utilities":
                action = "review usage and compare plans before the next billing cycle"
            else:
                action = "set a clear spending cap and review this category weekly"

            lines.append(
                f"- **{category}:** You spent ${amount:,.2f} ({percentage:.1f}%). "
                f"To save about ${potential_saving:,.2f}, {action}."
            )

    discretionary_percentage = float(analysis_data.get("discretionary_percentage", 0) or 0)
    discretionary_spending = float(analysis_data.get("discretionary_spending", 0) or 0)

    if discretionary_spending > 0:
        lines.append(
            f"- **Flexible spending check:** Coffee, dining and entertainment-style categories total "
            f"${discretionary_spending:,.2f} ({discretionary_percentage:.1f}%). "
            "Start with a small 5-10% reduction rather than cutting everything at once."
        )

    if total_refunds > 0:
        lines.append(
            f"- **Refund note:** You received ${total_refunds:,.2f} in refunds. "
            "Keep recording refunds separately so your budget reflects both spending behaviour and money returned."
        )

    lines.append("")
    lines.append("### Next step")
    lines.append(
        "Choose one category to improve this week, set a realistic dollar cap, "
        "and compare your next transaction export against this report."
    )
    lines.append("")
    lines.append("_This is budgeting education, not personal financial advice._")

    return "\n".join(lines)

# Test your function
recommendations = generate_financial_recommendations(analysis)
print(recommendations)
````
